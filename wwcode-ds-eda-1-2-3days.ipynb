# Tokenizing Text Data

```python
import nltk
from nltk.tokenize import word_tokenize

# Download the punkt tokenizer 
nltk.download('punkt')

# Assuming 'prompt' is the column containing text data
train_data['tokenized_prompt'] = train_data['prompt'].apply(lambda x: word_tokenize(str(x)))

# Display the first few rows to verify the tokenization
print(train_data[['prompt', 'tokenized_prompt']].head())
```

```python
# Convert the tokenized lists into strings
train_data['tokenized_prompt_str'] = train_data['tokenized_prompt'].apply(lambda x: ' '.join(x))

# Create a bag-of-words representation using TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=1000)  
X_text = vectorizer.fit_transform(train_data['tokenized_prompt_str'])
```

```python
from textblob import TextBlob

train_data['sentiment'] = train_data['prompt'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)
X_derived = train_data[['word_count', 'sentiment']]
```
